```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape") #to avoid encoding error, use unicode_escape
df.head()
```

```python
#Data Cleaning#
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True) #to drop the null columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace= True) #to drop the null values in particular columns
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

```python
df['Amount']=df['Amount'].astype('int') #astype is used to change the datatype of column
```

```python
df['Amount'].dtypes #used to check whether changes are done or not
```

```python
df.rename(columns={'Marital_Status': 'Shaadi'}) #if you want to rename the column name but will not get saved cuz inplace=True is not written
```

```python
#Exploratory Data Analysis
```

```python
df.columns
```

```python
#Gender
```

```python
ax= sns.countplot(x='Gender',data=df)
```

```python
ax= sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

#from below bar graph,we can say that-female buyers are more than male
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen) 
sns.set(rc={"figure.figsize":(3,2)})
```

```python
#Age
```

```python
sns.countplot(data=df, x='Age Group', hue='Gender')
```

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Amount',y='Age Group',data=sales_age, hue="Age Group") 


#from below we can see that most of the buyers are of age group 26-35 yrs
```

```python
#State
```

```python
#total no. of orders from top 12 states
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
```

```python
sns.barplot(x='State',y='Orders',data=sales_state)
sns.set(rc={'figure.figsize':(29,5)})
```

```python
#Total Amount/Top 6 states
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(6)
sns.barplot(x='State',y='Amount',data=sales_state)
sns.set(rc={'figure.figsize':(10,4)})
```

```python
plt.figure(figsize=(5,5))
plt.pie(sales_state['Amount'], labels=sales_state['State'], autopct='%1.1f%%', startangle=90)
plt.title('Sales Distribution by State')
plt.axis('equal')
plt.show()


#we can see that most of the orders/sales are from UttarPradesh,Maharashtra,Karnataka,Delhi....
```

```python
 #marital status
```

```python
ax = sns.countplot(x="Marital_Status", data = df)

sns.set(rc={'figure.figsize':(5,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(5,3)})
sns.barplot(data=sales_state,x="Marital_Status",y="Amount",hue="Gender")


#from below graph we can say that most of the buyers are women and they have high purchasing power
```

```python
#occupation
```

```python
ax= sns.countplot(x="Occupation",data=df)
sns.set(rc={"figure.figsize":(29,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending =False).head(10)
sns.set(rc={"figure.figsize":(29,9)})
sns.barplot(data=sales_state,x="Occupation",y="Amount")
```

```python
plt.figure(figsize=(14,7))


plt.scatter(sales_state['Amount'], sales_state['Occupation'], color="orange", s=200)


plt.grid(axis='x', linestyle='--', alpha=0.5)


plt.xlabel("Amount")
plt.ylabel("Occupation")
plt.title("Top 10 Occupations by Amount Spent")

plt.show()


#from below graph we can say that most of the buyers are working in IT sector,healthcare,aviation,banking.
```

```python
#product category
```

```python
ax= sns.countplot(x="Product_Category",data=df)
sns.set(rc={"figure.figsize":(29,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending =False).head(7)
sns.set(rc={"figure.figsize":(14,7)})
sns.barplot(data=sales_state,x="Product_Category",y="Amount")


#Top 7 most sold products
```
